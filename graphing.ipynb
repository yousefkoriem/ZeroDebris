```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('assets/data/ZeroDebris_Phase1_Data.csv')
df.head()
```

```python
plt.figure(figsize=(24, 8))
sns.scatterplot(data=df, x='ElapsedDays', y='Altitude')
plt.title('Altitude vs Elapsed Days', fontsize=16, fontweight='bold')
plt.xlabel('Elapsed Days', fontsize=14)
plt.ylabel('Altitude (km)', fontsize=14)
plt.xlim(0,2000)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='ElapsedDays', y='Altitude')
plt.title('Altitude vs Elapsed Days (Deorbit Phase)', fontsize=16, fontweight='bold')
plt.xlabel('Elapsed Days', fontsize=14)
plt.ylabel('Altitude (km)', fontsize=14)
plt.xlim(1800,1900)
plt.show()
```
